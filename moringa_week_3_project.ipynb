```python
#Dropping columns that are not necessary in our project

df.drop(['link'], axis = 1)
```

```python
#finding empty values in every column

df.isna().sum()
```

```python
#replacing null values

df['Bathrooms'].fillna(1,inplace=True)
```

```python
#checking unique values in the bedrooms column

df['Bedrooms'].unique()
```

```python
#checking the neighborhoods with most entries in the dataframe

df['Neighborhood'].value_counts()
```

Kilimani,Kileleshwa,westlands and lavington are the most advertised neighborhoods

```python
#checking the Agency with the most entries

df['Agency'].value_counts()
```

Prudential Green Ville ,Wealthlink Realtors Ltd ,Buy Rent Shelters,Pam Golding Properties  are the companies that run most of the marketing ads

```python
#confirming the number of bedrooms distribution 

df['Bedrooms'].value_counts()
```

```python
#Checking the number of bathroom that are common in the dataset


df['Bathrooms'].value_counts()
```

```python
#Removing the Ksh sign fron price column and converting it into float data type

df['Price'] = df['Price'].str.strip("KSh")
df['Price'].head()
```

```python
df.info()
```

```python
# converting the price column from object to integer

df['Price']=df['Price'].str.replace('[\,]',"").astype(int)
df['Price'].sum()
```
